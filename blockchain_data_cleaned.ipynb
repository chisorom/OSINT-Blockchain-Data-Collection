# Remove all \n

```python
#df2 = df.columns.str.replace('\n', '', regex=False).str.strip()
df = df.map(lambda x: x.replace('\n', '') if isinstance(x, str) else x)
#also replace question marks with Not known
df= df.map(lambda x: x.replace('?', 'Not known') if isinstance(x, str) else x)
df
```
